```python
import os
import glob
def generate_dataset(img_dir, xdim, ydim, nmax=5000) :
    
    label = 0
    label_names = []
    X = []
    y=[]
    for dirname in os.listdir(img_dir):
        print(dirname)
        label_names.append(dirname)
        data_path = os.path.join(img_dir + "/" + dirname,'*g')
        files = glob.glob(data_path)
        n=0
        for f1 in files:
            if n>nmax : break
            img = cv2.imread(f1) 
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
            img = cv2.resize(img, (xdim,ydim))
            X.append(np.array(img))
            y.append(label)
            n=n+1
        print(n,' images')
        label = label+1
    X = np.array(X)
    y = np.array(y)
    return X,y, label, label_names
```

```python
X_train,y_train, train_lb, train_lb_name = generate_dataset('./chest_xray/train/', 200, 200,1000)
```

```python
X_test,y_test, test_lb, test_lb_name = generate_dataset('./chest_xray/test/', 200, 200, 300)
```

```python
X_val,y_val, val_lb, val_lb_name = generate_dataset('./chest_xray/val/', 200, 200, 100)
```

```python
# normalize data
X_train = X_train/255
X_test = X_test/255
X_val = X_val/255
```

```python
X_train[0].shape
```

```python
# suffle data
from sklearn.utils import shuffle

X_train_shuffled, y_train_shuffled = shuffle(X_train, y_train, random_state=0)
X_test_shuffled, y_test_shuffled = shuffle(X_test, y_test, random_state=0)
X_val_shuffled, y_val_shuffled = shuffle(X_val, y_val, random_state=0)
```

```python
# MODEL
from tensorflow.keras import layers, models

data_augmentation = models.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal", input_shape=(200, 200,1)),
    layers.experimental.preprocessing.RandomRotation(0.15),
    layers.experimental.preprocessing.RandomZoom(0.13),
])

model = models.Sequential([
    data_augmentation,
    layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'), 
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.5),
    
    layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.5),
    
    layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.3),
    
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
    
])

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

##### Input an image to get the prediction

```python
from keras.preprocessing import image
import numpy as np

def img_preprocess(imgPath):
    img = image.load_img(imgPath, target_size=(200, 200))
    img = image.img_to_array(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = np.expand_dims(img, axis=0) # as model expect an array of images to predict
    img = img/255
    
    return img
```

```python
def get_label(prediction):
    if prediction > 0.5:
        return "Pneumonia"
    else:
        return "Normal"
```

```python
# normal lungs

pred = model.predict(img_preprocess('./chest_xray/test/NORMAL/IM-0003-0001.jpeg'))
print(f'Predicted probability :{pred[0][0]}')

print(f'Final result: {get_label(pred[0][0])}')

img = cv2.imread('./chest_xray/test/NORMAL/IM-0003-0001.jpeg')
plt.imshow(img)
plt.show()
```
